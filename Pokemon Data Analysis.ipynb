# 데이터 분석

## 1) 공격력이 가장 높은 포켓몬은?

```python
pokemon.sort_values(['Attack'], ascending=False).head() #공격력을 기준으로 정렬, 내림차순 정렬
```

## 2) 스피드가 가장 높은 포켓몬은?

```python
pokemon.sort_values(['Speed'], ascending=False).head() #속도를 기준으로 정렬, 내림차순 정렬
```

## 3) 전체 포켓몬들의 전반적인 Stats은?

```python
pokemon.describe().round(1) #pokemon 데이터의 기술통계량 계산
```

## 4) 전설 포켓몬은 일반포켓몬과 어떻게 다른가?

```python
GroupType1 = pokemon.groupby(['Legendary']) #Legendary를 기준으로 구분하고 GroupType1 변수로 지정
```

```python
GroupType1.mean().round(1) #GroupType1의 각 요소별 평균을 계산하고 소수 둘째자리에서 반올림
```

## 5) 전설 포켓몬은 누구인가?

```python
pokemon_Legen = pokemon.loc[pokemon['Legendary']==True] #loc을 이용해 전설 포켓몬만 선택
pokemon_Legen
```

## 6) 포켓몬 유형별 Stats 차이는?

```python
GroupType2 = pokemon.groupby(['Type 1']) #Type1을 기준으로 구분하고 GroupType2 변수로 지정
```

```python
GroupType2.mean().round(1)
```

## 7) 포켓몬 평가 지표 만들기

#### Defense per Attack Index

```python
pokemon.head() #포켓몬 데이터 확인
```

```python
pokemon['DAI'] = pokemon['Defense']/pokemon['Attack'] #DAI 지표를 정의하고 기존 필드에 추가
```

```python
pokemon.head() #포켓몬 데이터 확인
```

```python
pokemon['DAI'].describe() #DAI지표의 기술통계량 확인
```

```python
pokemon['DAI'].hist(bins=100, figsize=(10, 10)) #히스토그램으로 살펴보기
```

```python
pokemon_ver2 = pokemon.loc[pokemon['DAI']<5] #DAI값이 5이하인 것들만 정의해 poketmon_ver2로 정의
```

```python
pokemon_ver2['DAI'].hist(bins=100, figsize=(10, 10)) #히스토그램으로 살펴보기
```

## 8) 포켓몬 세부 속성별 상관관계 분석

```python
cor = pokemon.corr() #포켓몬 데이터 속성별 상관관계 분석
cor
```

# 데이터 시각화

```python
%matplotlib inline #시각화 자료를 주피터노트북에서 확인 가능
import seaborn as sns #seaborn 라이브러리 사용
import matplotlib.pyplot as plt #matplotlib 사용
```

```python
s = pokemon['Speed'] #포켓몬 데이터 중 Speed 필드를 저장
s.plot() 
```

```python
pokemon.boxplot(figsize=(10,5)) #포켓몬 데이터를 박스플롯으로 시각화
```

```python
pokemon.groupby('Type 1').boxplot(figsize=(30,30)) #그룹으로 구분된 시각화
```

```python
sns.catplot(
    x='Generation', 
    data=pokemon, 
    kind='count'    
).set_axis_labels('Generation', '# of Pokemon');
```

```python
sns.catplot(
    x='Generation',
    data=pokemon,
    col='Type 1',
    kind='count',
    col_wrap=3
).set_axis_labels('Generation', '# of Pokemon');
```

```python
sns.lmplot(x='Attack',y='Defense',data=pokemon,hue='Type 1')
plt.ylim(0,None)
```

```python
sns.heatmap(cor)
```
